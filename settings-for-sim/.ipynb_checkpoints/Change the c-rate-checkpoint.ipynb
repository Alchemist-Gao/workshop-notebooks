```python
import pybamm
```

```python
model = pybamm.lithium_ion.SPM()
sim = pybamm.Simulation(model)
sim.solve()
```

```python
model = pybamm.lithium_ion.SPM()
sim_2C = pybamm.Simulation(model, C_rate=2)
sim_2C.solve()
```

```python
plot = pybamm.QuickPlot([sim.solution, sim_2C.solution], output_variables=["Terminal voltage [V]"], labels=["1C", "2C"])
plot.plot(0)
```
